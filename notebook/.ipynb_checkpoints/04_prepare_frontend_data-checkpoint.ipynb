```python
import os
import shutil

SOURCE_DIR = r"D:\PROJECT\UIDAI hackathon\notebook\outputs"
TARGET_DIR = r"D:\PROJECT\UIDAI hackathon\outputs"

os.makedirs(TARGET_DIR, exist_ok=True)

for file in os.listdir(SOURCE_DIR):
    src = os.path.join(SOURCE_DIR, file)
    dst = os.path.join(TARGET_DIR, file)

    if os.path.isfile(src):
        shutil.move(src, dst)
        print(f"Moved: {file}")
```

```python
import pandas as pd

for file in os.listdir(OUTPUT_DIR):
    if file.endswith(".csv"):
        csv_path = os.path.join(OUTPUT_DIR, file)
        json_path = os.path.join(
            FRONTEND_DATA_DIR,
            file.replace(".csv", ".json")
        )

        df = pd.read_csv(csv_path)

        df.to_json(
            json_path,
            orient="records",
            indent=2
        )

        print(f"Converted {file}")
```

```python
import pandas as pd
import os

OUTPUT_DIR = r"D:\PROJECT\UIDAI hackathon\outputs"
FRONTEND_DATA_DIR = r"D:\PROJECT\UIDAI hackathon\frontend\public\data"

os.makedirs(FRONTEND_DATA_DIR, exist_ok=True)

for file in os.listdir(OUTPUT_DIR):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(OUTPUT_DIR, file))
        df.to_json(
            os.path.join(FRONTEND_DATA_DIR, file.replace(".csv", ".json")),
            orient="records",
            indent=2
        )
        print(f"Converted {file}")
```
